```python
#!pip install -U pandasql
from pandasql import sqldf
import os
import json
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import numpy as np
import re
```

```python
founders_data = pd.read_csv('/Users/chaosefat/TUM/IDP/Visualization/founders_dataset_IDP.csv')
```

```python
json_directory = '/Users/chaosefat/TUM/IDP/Visualization/all/json/'
i = 0
ocean_personalities = ['openness','conscientiousness','extraversion','agreeableness','neuroticism']
#founders_data[ocean] = 'N/A'
name_list=list();
opennes_list=list()
for ocean in ocean_personalities:
    for json_file_path in os.listdir(json_directory):
        file_peson_name = json_file_path.rstrip(".json")
        file_peson_name = re.sub('([A-Z])', r' \1', file_peson_name)
        file_peson_name = file_peson_name.strip()
        name_list.append(file_peson_name)

        json_file = open(json_directory + json_file_path,
                         'r', encoding='utf-8')
        content = json.load(json_file)

        if "results" in content:

            content2 = content['results'][0]

            if "personality" in content2:

                score = content2['personality'][ocean]
                #opennes_list.append(score)
                founders_data.loc[founders_data['person_name'] == file_peson_name, ocean] = score

            else:
                continue

        else:
            continue
        
    
```
